```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt\
# some basic libraries required to do multivariate regression
```

```python
df=pd.read_csv('data.csv')
df
#importing the sample data from a csv file 
```

```python
# y = m1*weekday + m2*workingday + m3*weathersit + c
reg = linear_model.LinearRegression()
# weekday , workingday , weathersit are independent variable . cnt is a dependent variable 
reg.fit(df[['weekday','workingday','weathersit']],df.cnt)
# training of model complete
```

```python
reg.coef_
# will give three values of m1 , m2 , m3 respectively
```

```python
reg.intercept_
# will give the value of c i.e. intercept
```

```python
reg.predict([[2,1,1]])
```

```python
reg.predict([[2,0,1]])
```
